```python
import fnmatch, os
full_path = os.path.join(path, evaluation)

def replace (key, replace, wildcard = '*'):

  for file in os.listdir( full_path ):
    if fnmatch.fnmatch(file, ''.join( [wildcard, key, wildcard] ) ):
      os.rename(os.path.join(full_path, file), os.path.join(full_path, file.replace(key, replace)) )
```

```python
replacements1 = [':', '  ', 'v0.0.1', 'v0.0.2', 'v.0.2', ' Structured ', 'Interview', 'v0.1', 'v0.2', 'Evaluation', '-', 'v0.3', '  ', 'v1.0.0', 'evaluation']
replacements2 = [' .', '..']
replacements3 = ['AWS', 'Kubernetes', 'Terraform', 'Serverless', 'Azure', 'Caylent']
replacements4 = [
  # removed line
                ]
for item in replacements1:
  replace(key = item, replace = '')

for item in replacements2:
  replace(key = item, replace = '.')

for item in replacements3:
  replace(key = item, replace = ' - {}'.format(item))

for item in replacements4:
  replace(key = item[0], replace = item[1])
```

```python
import binascii
accum_hash = [] ###### hashing
df = pd.DataFrame() # create df
for file in os.listdir( full_path ): # for every file in the folder
  hash = str(binascii.hexlify(os.urandom(3)).decode("utf-8")) # generates a random hash
  employee = file[:file.find('-') - 1] # takes the employee name
  if fnmatch.fnmatch(file, '*-*' ) and employee not in [x[1] for x in accum_hash]: # validates if employee already has a hash
    accum_hash.append ([hash, employee]) # creates matrix for hash and employee 
    os.rename(os.path.join(full_path, file), os.path.join(full_path, file.replace(employee, hash)) ) # rename the file to format (hash - employee name.xlsx)
    df = pd.DataFrame (accum_hash) # create df from matrix
    df.rename(columns = {0:'Hash', 1:'Employee'}, inplace = True) # rename columns
    # df['Counts'] = df.groupby(['Hash'])['Employee'].transform('count') # generates the column baseline
  elif fnmatch.fnmatch(file, '*-*' ) and employee in [x[1] for x in accum_hash]: # if employee already has a hash
    hash = (df.loc[df.Employee == employee]['Hash']).values[0] # takes the hash
    os.rename(os.path.join(full_path, file), os.path.join(full_path, file.replace(employee, hash)) ) # rename the file to format (hash - employee name.xlsx)
    
csv_file = ('All {} hashed employees.csv' .format(str(df.shape[0]))) # csv
df.to_csv(path+csv_file, sep=',', index = False) # saves csv
df.head(50)
```
